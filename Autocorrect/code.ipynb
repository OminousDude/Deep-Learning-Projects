```python
def load_dictionary(file_path):
    with open(file_path, 'r') as file:
        return [line.strip() for line in file]
```

```python
def get_distance(str1, str2):
    ind_matrix = [[i] for i in range(1, str1.__len__() + 1)]
    ind_matrix.insert(0, list(range(0, str2.__len__() + 1)))
    for j in range(1, str2.__len__() + 1):
        for i in range(1, str1.__len__() + 1):
            add_distance = None
            if str1[i - 1] == str2[j - 1]:
                add_distance = 0
            else:
                add_distance = 1
            ind_matrix[i].insert(j, min(ind_matrix[i - 1][j] + 1,
                               ind_matrix[i][j - 1] + 1,
                               ind_matrix[i - 1][j - 1] + add_distance))
    return ind_matrix[-1][-1]
```

```python
def spell_check(word, dictionary):
    suggestions = []

    for correct_word in dictionary:
        distance = get_distance(word, correct_word)
        suggestions.append((correct_word, distance))

    suggestions.sort(key=lambda x: x[1])
    return suggestions[:10000]
```

```python
dictionary = load_dictionary("words.txt")
misspelled_word = "float"
suggestions = spell_check(misspelled_word, dictionary)
for word, distance in suggestions:
    print(f"{word} (Distance: {distance})")
```
